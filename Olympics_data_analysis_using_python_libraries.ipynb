**Differences spotted between boxplot created by matplotlib and seaborn**:

Seaborn simplifies box plot creation with built-in grouping (via hue) and better default aesthetics, while matplotlib requires more manual effort for similar plots.

Seaborn integrates seamlessly with Pandas DataFrames for easy plotting, whereas matplotlib often requires data extraction.

Matplotlib offers more control but less convenience compared to Seaborn's user-friendly interface for box plots.

For Swimming as the sport finding the top 3 participating nations and creating a separate dataset based on this.

```python
swimming_data = olympics_data[olympics_data['Sport'] == 'Swimming']
team_count = swimming_data.groupby('Team')['Team'].count()
top_3_nation = team_count.sort_values(ascending = False).head(3)
top_3_nation
```

**Compare the average Height based on gender**

```python
top_3_swimming_data = swimming_data[swimming_data['Team'].isin(top_3_nation.index)]
sns.set_style('dark', {'axes.grid': False})
plt.figure(figsize = (4,5))
bar = sns.barplot(x = top_3_swimming_data['Sex'], y = top_3_swimming_data['Height'], errorbar = None, hue = top_3_swimming_data['Sex'], legend = True)
plt.title('Height of Top 3 Swimming Teams')
plt.tight_layout()
plt.show()
```

**Compare the average Weight based on gender**

```python
sns.set_style('dark', {'axes.grid': False})
plt.figure(figsize = (4,5))
sns.barplot(x = top_3_swimming_data['Sex'], y = top_3_swimming_data['Weight'], errorbar = None, hue = top_3_swimming_data['Sex'], legend = True)
plt.title('Height of Top 3 Swimming Teams')
plt.show()
```

**Conclusion:**

From the line plot, which displays the relationship between year and height for Olympic athletes based on gender, the following inferences can be made:

1. **Overall Increase in Height Over Time:**
   - Both male (M) and female (F) athletes' heights show an increasing trend from the 1960s to the 2010s. This suggests a general increase in the average height of Olympic athletes over time.

2. **Males Are Generally Taller Than Females:**
   - Throughout the years, male athletes (M) consistently have higher average heights than female athletes (F). This is indicated by the orange line (M) being higher than the blue line (F) across all years.

3. **Greater Variability in Female Heights:**
   - The line for female athletes (F) exhibits more fluctuations after the year 1990, indicating higher variability in the heights of female athletes in that period. This could be due to changes in the types of sports, events, or participants in the Olympics for women.

4. **Male Heights Stabilize:**
   - For male athletes (M), the height shows a steadier increase without much fluctuation compared to the females. After around 1990, the growth in male athletes' heights seems to have slowed down, stabilizing around 180 cm.

5. **Height Plateaus in Recent Years:**
   - Both lines suggest that the increase in height is slowing down, with only small increases or slight fluctuations in height in the most recent years. This could imply a plateau in the average height of Olympic athletes.

In summary, the plot suggests an increasing trend in athlete height over time, with male athletes generally being taller than female athletes and a greater variability in female athlete heights after 1990.

**Creating a sentiment column athletes who won 1 medal were having neutral sentiment, 0 negative and more than 2 positive**

```python
import numpy as np

condition = [(olympics_data['Medal'] > 1), (olympics_data['Medal'] == 1), (olympics_data['Medal'] == 0)]
values = ['Positive', 'Neutral', 'Negative']
olympics_data['Sentiment'] = np.select(condition, values)
olympics_data
```

**Creating a Pie Chart**

```python
plt.figure(figsize = (5,5))
plt.pie(olympics_data['Sentiment'].value_counts(), labels = ['Negative', 'Positive', 'Neutral'], autopct = '%1.1f%%')
plt.title('Distribution of Sentiment')
plt.show()
```

**Creating a customized scatterplot of Height and Weight**

```python
plt.figure(figsize = (8,5))
plt.title('Relation between Height and Weight')
plt.scatter(x = 'Height', y = 'Weight', data = olympics_data, marker = 'o', color = 'Yellow', edgecolor = 'Orange', s = 50)
plt.xlabel('Height')
plt.ylabel('Weight')
plt.show()
```

**Further customize the plot by colour coding and also extract the positive piece by 30% away from the complete chart.**

```python
import matplotlib.pyplot as plt
plt.figure(figsize = (6,6))
plt.pie(olympics_data['Sentiment'].value_counts(), labels = ['Negative', 'Positive', 'Neutral'], autopct = '%1.1f%%', explode = (0,0.3,0), colors = ['r','g','b'])
plt.title('Distribution of Sentiment')
```
